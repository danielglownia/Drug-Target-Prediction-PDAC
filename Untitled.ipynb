```python
expression_dict = {}
for x in range(27):
    threshold_value = x

    mask = (df > threshold_value).all(axis=1)

    # Use the mask to filter the DataFrame
    result_df = df[mask]
    expression_dict[x] = result_df.shape[0]
```

```python
exp_df_level = pd.DataFrame(expression_dict.items(), columns=['exp_level', 'num_genes'])
```
